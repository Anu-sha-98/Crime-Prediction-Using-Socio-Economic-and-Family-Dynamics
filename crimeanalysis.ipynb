### Remove Non-predictive variables

```python
df = df.drop(columns=['Êcommunityname', 'countyCode', 'communityCode', 'fold'], axis=1)
df.head()
```

```python
from pandas import DataFrame

df = df.replace('?', np.nan)
missing = df.columns[df.isnull().any()]
missing
```

```python
missing.shape
```

```python
df[missing[0:13]].describe()
```

```python
df[missing[13:26]].describe()
```

```python
for col in df[missing] :
    print(col,df[col].isnull().sum())
```

**Missing value imputation with mean**

```python
from sklearn.impute import SimpleImputer
import numpy as np

for col in missing :
    imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
    imputer = imputer.fit(df[[col]])
    df[[col]] = imputer.transform(df[[col]])
    print(df[col].isnull().any())
```
